```python
# Susceptibilidad antifngica individual
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Crear el dataframe con los datos de susceptibilidad (simplificado para 3 antifngicos)
data = {
    'Especie': [
        'C. albicans', 'C. dublinensis', 'C. famata', 'C. glabrata', 'C. krusei',
        'C. parasilosis', 'C. tropicalis', 'G. capitatum', 'G. candidum', 'T. asahii', 'Cultivos mixtos'
    ],
    'Ketoconazol_S': [44.8, 100, 100, 81.3, 100, 100, 100, 100, 100, 100, 60],
    'Ketoconazol_I': [55.2, 0, 0, 10.4, 0, 0, 0, 0, 0, 0, 20],
    'Ketoconazol_R': [0, 0, 0, 8.3, 0, 0, 0, 0, 0, 0, 20],

    'Fluconazol_S': [4.5, 0, 100, 75, 50, 100, 0, 66.7, 0, 100, 40],
    'Fluconazol_I': [0, 0, 0, 4.2, 0, 0, 0, 0, 0, 0, 0],
    'Fluconazol_R': [95.5, 100, 0, 20.8, 50, 0, 100, 33.3, 100, 0, 60],

    '5FC_S': [19.4, 25, 100, 56.3, 37.5, 0, 0, 0, 0, 0, 0],
    '5FC_I': [29.9, 0, 0, 14.6, 0, 0, 0, 0, 0, 0, 0],
    '5FC_R': [50.7, 75, 0, 29.2, 62.5, 100, 100, 100, 100, 100, 100],
}

df = pd.DataFrame(data)

def analyze_antifungal(df, antifungal_prefix):
    table = df[[f"{antifungal_prefix}_S", f"{antifungal_prefix}_I", f"{antifungal_prefix}_R"]].to_numpy()
    chi2, p, dof, expected = chi2_contingency(table)
    n = table.sum()
    r, c = table.shape
    v_cramer = np.sqrt(chi2 / (n * (min(r, c) - 1)))
    residuals = (table - expected) / np.sqrt(expected)

    print(f"--- {antifungal_prefix} ---")
    print(f"Chi = {chi2:.2f}")
    print(f"Grados de libertad = {dof}")
    print(f"p-valor = {p:.5e}")
    print(f"V de Cramer = {v_cramer:.3f}")
    print("\nResiduos estandarizados:")
    print(pd.DataFrame(residuals, index=df['Especie'], columns=["S", "I", "R"]))
    print("\n")

# Analizar cada antifngico
analyze_antifungal(df, 'Ketoconazol')
analyze_antifungal(df, 'Fluconazol')
analyze_antifungal(df, '5FC')
```

```python
# Distribucin de comorbilidades
import numpy as np
import scipy.stats as stats
import math

# Conteos aproximados de comorbilidades (sumando 155)
observed = np.array([54, 26, 9, 7, 7, 15, 5, 32]).astype(int) # Explicitly convert to integers

# Total de casos (calculated from the sum of observed)
n = observed.sum()

# Esperados bajo hiptesis de distribucin uniforme
expected = np.full_like(observed, n / observed.size, dtype=float) # Ensure expected is float for division

# 1. Chi-cuadrado de bondad de ajuste
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# 2. Grados de libertad
dof = observed.size - 1

# 3. V de Cramer
phi2 = chi2_stat / n
v_cramer = math.sqrt(phi2 / dof)

# Resultados
print(f"Chi = {chi2_stat:.2f}")
print(f"p-valor = {p_value:.2e}")
print(f"Grados de libertad = {dof}")
print(f"V de Cramer = {v_cramer:.2f}")
```
